## 1.2 Data quality

"Meanwhile, the DarkCideS 1.0 dataset contains occurrence data for 402 species from 16 families representing 59% of all cave-dwelling species"
The actual survey coverage problem - only a little more than half of the species data were collected

"at least 679 species are known to be cave-dwelling"
Theoretical population size - a complete species list based on IUCN expert judgment

1. Incomplete data collection:
   - IUCN lists 679 theoretical cave bat species
   - DarkCideS only collected actual cave records of 402 species
   - The coverage rate is only 59%

2. Geographical bias:
   - The document mentions the existence of "geographical biases"
   - Data is lacking in some regions, such as tropical Africa
   - Indonesia is expected to have many cave bat species, but there are few records in the database

3. Different data sources:
   - IUCN based on expert judgment and literature review
   - DarkCideS is based on actual field survey records

So the observed data are inconsistent (Cave=0 but 71 species appear in the cave, Cave=1 but 280 species are not found in the cave) for this reason: the actual survey coverage is incomplete and only captures part of the theoretical expectations.

"We found a marginally significant relationship between the species richness and proportion of threatened species between the IUCN-based global cave-dwelling bat and DarkCideS datasets (Kendall's τ b = 0.60, P = 0.07)."
"We found a marginally significant relationship between the species richness and proportion of threatened species between the IUCN-based global cave-dwelling bat dataset and DarkCideS dataset"

1. Data quality issues: Although DarkCideS only collected 59% of the species data, this 59% sample is statistically representative, not randomly missing, but a systematic representative sample.
2. Avoid sample bias: Prove that the 402 species collected are not a specific type of bat, but a smaller version of the whole
3. Effective conservation priorities: Conservation decisions based on these 402 species can be reasonably extended to all 679 species
4. Research conclusions are credible: The scientific conclusions drawn from this incomplete data set are valid for the entire cave bat population

```python
from scipy.spatial import cKDTree
import numpy as np

# Filter parameters
threshold = 0.0001
species_in_taxonomic = set(bat_taxonomic['Species.name'])

print("=== Species Loss Analysis ===\n")

# Step 1: Basic filtering
step1_filtered = bat_cave[
    bat_cave['Species.name'].isin(species_in_taxonomic) &
    ~pd.isna(bat_cave['Latitude']) &
    ~pd.isna(bat_cave['Longitude'])
]
step1_species = set(step1_filtered['Species.name'])

# Step 2: Spatial matching
cave_coords = cave_surroundings[['Longitude', 'Latitude']].values
tree = cKDTree(cave_coords)

filtered_rows = []
for idx, row in step1_filtered.iterrows():
    neighbors = tree.query_ball_point([row['Longitude'], row['Latitude']], threshold)
    
    for neighbor_idx in neighbors:
        cave_row = cave_surroundings.iloc[neighbor_idx]
        if (cave_row['Biogeographical.realm'] == row['Biogeographical.realm'] and
            cave_row['Country'] == row['Country.record'] and
            cave_row['Cave_Name'] == row['Cave.site']):
            filtered_rows.append(idx)
            break

# Results
filtered_bat_cave = bat_cave.loc[filtered_rows]
step2_species = set(filtered_bat_cave['Species.name'])
lost_species = step1_species - step2_species

# Print results
print(f"Step 1 (taxonomic + coordinates): {len(step1_filtered)} records, {len(step1_species)} species")
print(f"Step 2 (after spatial matching): {len(filtered_bat_cave)} records, {len(step2_species)} species")
print(f"Record loss: {(len(step1_filtered)-len(filtered_bat_cave))/len(step1_filtered)*100:.1f}%")
print(f"Species loss: {len(lost_species)/len(step1_species)*100:.1f}%")
print(f"Lost species count: {len(lost_species)}")

if lost_species:
    print("\nLost species:")
    for species in sorted(lost_species):
        count = len(step1_filtered[step1_filtered['Species.name'] == species])
        print(f"  - {species}: {count} records")

print(f"\nFinal result: {len(filtered_rows)} rows ({len(filtered_rows)/len(bat_cave)*100:.1f}% of total)")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Prepare data
original_taxonomic = bat_taxonomic[bat_taxonomic['Species.name'].isin(bat_cave['Species.name'])]
matched_taxonomic = bat_taxonomic[bat_taxonomic['Species.name'].isin(filtered_bat_cave['Species.name'])]

variables = ['Population.status', 'Conservation.status', 'Extinction.risk']
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, var in enumerate(variables):
    orig_dist = original_taxonomic[var].value_counts(normalize=True)
    match_dist = matched_taxonomic[var].value_counts(normalize=True)
    
    categories = sorted(set(orig_dist.index) | set(match_dist.index))
    orig_values = [orig_dist.get(cat, 0) for cat in categories]
    match_values = [match_dist.get(cat, 0) for cat in categories]
    
    # Plot comparison
    x = np.arange(len(categories))
    axes[i].bar(x - 0.2, orig_values, 0.4, label='Original', alpha=0.7)
    axes[i].bar(x + 0.2, match_values, 0.4, label='Matched', alpha=0.7)
    axes[i].set_title(var)
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(categories, rotation=45, ha='right')
    axes[i].legend()
    
    # Chi-square test
    orig_counts = original_taxonomic[var].value_counts()
    match_counts = matched_taxonomic[var].value_counts()
    contingency_table = [[orig_counts.get(cat, 0), match_counts.get(cat, 0)] for cat in categories]
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    print(f"{var}: χ² = {chi2:.3f}, p = {p_value:.3f}")

plt.tight_layout()
plt.show()
```

```python
# Biogeographical realm analysis
original_biogeo = bat_cave['Biogeographical.realm'].value_counts(normalize=True)
matched_biogeo = filtered_bat_cave['Biogeographical.realm'].value_counts(normalize=True)

# Plot
regions = sorted(set(original_biogeo.index) | set(matched_biogeo.index))
orig_values = [original_biogeo.get(region, 0) for region in regions]
match_values = [matched_biogeo.get(region, 0) for region in regions]

x = np.arange(len(regions))
plt.figure(figsize=(12, 6))
plt.bar(x - 0.2, orig_values, 0.4, label='Original', alpha=0.7)
plt.bar(x + 0.2, match_values, 0.4, label='Matched', alpha=0.7)

plt.xlabel('Biogeographical Realm')
plt.ylabel('Proportion')
plt.title('Biogeographical Realm Distribution Comparison')
plt.xticks(x, regions, rotation=45, ha='right')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print differences
for region in regions:
   orig_pct = original_biogeo.get(region, 0) * 100
   match_pct = matched_biogeo.get(region, 0) * 100
   diff = match_pct - orig_pct
   print(f"{region}: {orig_pct:.1f}% vs {match_pct:.1f}% (diff: {diff:+.1f}%)")
```

```python
from scipy.spatial import cKDTree
import numpy as np

# Rename duplicate columns
bat_taxonomic_renamed = bat_taxonomic.rename(columns={
    col: f"{col}_taxonomic" for col in bat_taxonomic.columns 
    if col in filtered_bat_cave.columns and col != 'Species.name'
})

cave_surroundings_renamed = cave_surroundings.rename(columns={
    col: f"{col}_surroundings" for col in cave_surroundings.columns 
    if col in filtered_bat_cave.columns
})

# Merge bat_cave and bat_taxonomic
merged_bat_cave = pd.merge(filtered_bat_cave, bat_taxonomic_renamed, on='Species.name', how='left')

# Build spatial index
threshold = 0.0001
coords = cave_surroundings_renamed[['Longitude_surroundings', 'Latitude_surroundings']].values
tree = cKDTree(coords)

final_data = []
for idx, row in merged_bat_cave.iterrows():
    row_dict = row.to_dict()
    
    if not (pd.isna(row['Latitude']) or pd.isna(row['Longitude'])):
        neighbors = tree.query_ball_point([row['Longitude'], row['Latitude']], threshold)
        
        for neighbor_idx in neighbors:
            cave_row = cave_surroundings_renamed.iloc[neighbor_idx]
            if (cave_row['Biogeographical.realm_surroundings'] == row['Biogeographical.realm'] and
                cave_row['Country'] == row['Country.record'] and
                cave_row['Cave_Name'] == row['Cave.site']):
                
                # Merge environmental data
                for col in cave_surroundings_renamed.columns:
                    if col not in ['Latitude_surroundings', 'Longitude_surroundings']:
                        row_dict[col] = cave_row[col]
                break
    
    final_data.append(row_dict)

merged_bat_cave = pd.DataFrame(final_data)

print(f"Merged dataset: {len(merged_bat_cave)} records")
print(f"Number of columns: {len(merged_bat_cave.columns)}")
print(f"Original bat_cave: {len(bat_cave)} records")
```

```python
import numpy as np
from scipy.spatial import cKDTree

# Prepare data
print(f"Original data: {len(merged_bat_cave)} records")
valid_data = merged_bat_cave.dropna(subset=['Latitude', 'Longitude', 'Species.name']).reset_index(drop=True)
print(f"Valid data: {len(valid_data)} records")

coords = valid_data[['Longitude', 'Latitude']].values
species = valid_data['Species.name'].values
tree = cKDTree(coords)

# Remove duplicates within 10m threshold
threshold_deg = 10 / 111000
to_remove = set()

for i, (coord, spec) in enumerate(zip(coords, species)):
    if i in to_remove:
        continue
    for j in tree.query_ball_point(coord, threshold_deg * np.sqrt(2)):
        if i < j and species[i] == species[j] and max(abs(coords[i] - coords[j])) <= threshold_deg:
            to_remove.add(j)

# Update dataset
merged_bat_cave_dedup = valid_data.iloc[[i for i in range(len(valid_data)) if i not in to_remove]].reset_index(drop=True)

print(f"Deduplicated data: {len(merged_bat_cave_dedup)} records")
print(f"Removed {len(to_remove)} duplicate records")
print(f"Deduplication rate: {len(to_remove)/len(valid_data)*100:.1f}%")
print(f"Species before: {len(valid_data['Species.name'].unique())}")
print(f"Species after: {len(merged_bat_cave_dedup['Species.name'].unique())}")
```

```python
from scipy.spatial import cKDTree
import numpy as np

# Set 5-meter radius
radius = 5
radius_deg = radius / 111000

# Prepare data
coords = merged_bat_cave_dedup[['Longitude', 'Latitude']].values
cave_sites = merged_bat_cave_dedup['Cave.site'].values
countries = merged_bat_cave_dedup['Country.record'].values
tree = cKDTree(coords)

# Find cave groups
assigned = np.zeros(len(coords), dtype=bool)
caves = []

for i in range(len(coords)):
   if assigned[i]:
       continue
   neighbors = tree.query_ball_point(coords[i], radius_deg)
   for j in neighbors:
       assigned[j] = True
   caves.append(neighbors)

# Analyze country-inconsistent cave groups
country_inconsistent_caves = []
for cave_idx, cave in enumerate(caves):
   if len(cave) > 1:
       country_values = [countries[idx] for idx in cave]
       unique_countries = set(country_values)
       if len(unique_countries) > 1:
           country_inconsistent_caves.append({
               'cave_group': cave_idx,
               'point_indices': cave,
               'countries': country_values,
               'unique_countries': unique_countries,
               'cave_sites': [cave_sites[idx] for idx in cave],
               'point_count': len(cave)
           })

print(f"Country inconsistency analysis (radius {radius}m):")
print(f"Total cave groups: {len(caves)}")
print(f"Country-inconsistent groups: {len(country_inconsistent_caves)}")
print(f"Inconsistency rate: {len(country_inconsistent_caves)/len(caves)*100:.1f}%")

# Show details for all inconsistent groups
print(f"\nCountry-inconsistent cave group details:")
for i, cave_info in enumerate(country_inconsistent_caves):
   print(f"\n--- Cave Group {i+1} ---")
   print(f"Point count: {cave_info['point_count']}")
   print(f"Countries: {list(cave_info['unique_countries'])}")
   print(f"Cave names: {list(set(cave_info['cave_sites']))}")
   
   for j, point_idx in enumerate(cave_info['point_indices']):
       lat = coords[point_idx][1]
       lon = coords[point_idx][0]
       country = countries[point_idx]
       site = cave_sites[point_idx]
       print(f"  Point{j+1}: {country} - {site} (lat:{lat:.6f}, lon:{lon:.6f})")

# Country combination statistics
country_combinations = {}
for cave_info in country_inconsistent_caves:
   countries_tuple = tuple(sorted(cave_info['unique_countries']))
   if countries_tuple not in country_combinations:
       country_combinations[countries_tuple] = []
   country_combinations[countries_tuple].append(cave_info)

print(f"\nCountry combination statistics:")
for countries_combo, caves_list in sorted(country_combinations.items(), key=lambda x: len(x[1]), reverse=True):
   print(f"{countries_combo}: {len(caves_list)} cave groups")
   for cave_info in caves_list[:3]:
       sites = list(set(cave_info['cave_sites']))
       print(f"  Example: {sites}")
   if len(caves_list) > 3:
       print(f"  ... and {len(caves_list)-3} more")
```
